```python
import pandas as pd

df = pd.read_excel("index.xlsx", sheet_name="Sheet1")
df.columns = df.columns.str.strip()

def fix_date(x):
    if pd.isna(x):
        return ""

    try:
        # handles datetime objects + most strings
        return pd.to_datetime(x, dayfirst=True).strftime("%d-%m-%Y")
    except:
        try:
            # fallback for dot-separated weird strings
            x = str(x).strip()
            parts = x.split(".")
            
            if len(parts) == 3:
                day = parts[0].zfill(2)
                month = parts[1].zfill(2)
                year = parts[2]

                if len(year) == 2:
                    year = "20" + year

                return f"{day}-{month}-{year}"
        except:
            pass

        return ""

df["DATE"] = df["DATE"].apply(fix_date)

display(df)
```

```python
with pd.ExcelWriter(
    "index.xlsx", 
    mode="r+", 
    engine="openpyxl",
    if_sheet_exists="replace",
    datetime_format="DD-MM-YYYY") as writer:
    df.to_excel(writer, sheet_name="date sheet", index=False)
```
